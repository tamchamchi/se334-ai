````python
from pyngrok import ngrok
import uvicorn
import threading

# Mở cổng 8000 public
public_url = ngrok.connect(8000)
print("🔥 Public URL:", public_url)

# Chạy FastAPI trên cổng 8000 ở background
def run_api():
    uvicorn.run("api.main:app", host="0.0.0.0", port=8000)

threading.Thread(target=run_api).start()
````

# 🧪 How to Test the Sentiment Analysis API

This guide shows how to test the FastAPI sentiment analysis API using Swagger UI and Postman.

---

## ✅ 1. Test via Swagger UI

Once your API is running via ngrok, open the Swagger UI:

🔗 `https://<your-ngrok-url>.ngrok-free.app/docs`

- Click `POST /predict`
- Click **"Try it out"**
- Enter sample input:

```json
{
  "texts": [
    "I love this product!",
    "The service was terrible.",
    "Nothing special about today."
  ]
}
```
## ✅ 2. Test via Postman
🔧 Setup:
Method: POST

URL: https://<your-ngrok-url>.ngrok-free.app/predict

Headers:

Content-Type: application/json

🧾 Body (raw JSON):
```json
{
  "texts": [
    "Dịch vụ rất tốt, tôi hài lòng.",
    "Thật tệ, tôi sẽ không quay lại.",
    "Thời tiết hôm nay thật bình thường."
  ]
}
```
Click Send to get results.
